Filtering dataframe on a condition

```python
mask = data['city'] == 'Hyderabad'
data[mask]
```

```python
data[mask].shape
```

```python
#making a function
def get_city_match_count(city) :
    mask = data['city'] == city
    return data[mask].shape
get_city_match_count('Hyderabad')    
```

```python
#Improved
mask1 = data['city'] =='Hyderabad'
mask2 = data['date'] > '2017-04-09'
data[mask1 & mask2].shape[0]
```
